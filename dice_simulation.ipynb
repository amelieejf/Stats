```python
# @title
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize data
results = []

# Function to simulate dice rolls
def simulate_dice_rolls(num_rolls):
    global results
    rolls = np.random.randint(1, 7, size=(num_rolls, 2))
    sums = np.sum(rolls, axis=1)
    results.extend(sums)
    update_plot()

# Function to update the plot
def update_plot():
    clear_output(wait=True)  # Clear previous output
    display(widgets.HBox([button_roll_once, button_roll_10, button_roll_100]))  # Redisplay buttons
    plt.figure(figsize=(10, 6))
    plt.hist(results, bins=np.arange(2, 14) - 0.5, edgecolor='black', rwidth=0.8)
    plt.title("Sum of Two Dice Rolls")
    plt.xlabel("Sum")
    plt.ylabel("Frequency")
    plt.xticks(range(2, 13))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Button creation
button_roll_once = widgets.Button(description="Roll Once")
button_roll_10 = widgets.Button(description="Roll 10 Times")
button_roll_100 = widgets.Button(description="Roll 100 Times")

# Button event handlers
def roll_once(_):
    simulate_dice_rolls(1)

def roll_10(_):
    simulate_dice_rolls(10)

def roll_100(_):
    simulate_dice_rolls(100)

button_roll_once.on_click(roll_once)
button_roll_10.on_click(roll_10)
button_roll_100.on_click(roll_100)

# Display buttons and initial plot
display(widgets.HBox([button_roll_once, button_roll_10, button_roll_100]))
update_plot()
```
